## This application is to train Isolation forest to get the labels and finally train a random forest using these labels.

### This folder contains following 5 files
##### * Variables  -- This file has all the imports and variables.
##### * Driver_function -- This is the main file which will execute all the commands or files.
##### * 1.Load_Data_And_Construct_Traces  -- Here the raw data is loaded, traces as extracted and three phases of the traces are marked.
##### * 2.Feature_Creation -- This file creates features from the traces.
##### * 3.Training_Isolation_Forest_And_Random_Forest -- Here the features are used to train Isolation forest and finally train a Random forest using             the labels from Isolation forest.
