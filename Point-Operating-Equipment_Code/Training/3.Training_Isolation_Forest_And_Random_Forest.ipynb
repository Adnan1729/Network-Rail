##### Training_Isolation_forest_and_random_forest

```
def Train_Isolation_forest_and_Random_Forest(Asset_list,Raw_data_path,RN,NR,features_to_test,Model_Output_Path):
  '''This function takes the features as input that will be training data for Isolation forest. First we train Isolation forest that labels all the traces as -1 meaning anomalous or 1 meaning Normal. Now these labels will be used to train Random forest. Finally, the trained random forest is saved which can be used to predict on new dataset
  
  Input :: Pandas dataframe containing all seven features for all the traces
  Output :: Trained random forest model in .sav format'''
  
  Feature_pdf_IF = create_features(Asset_list,Raw_data_path,RN,NR)
  ### calling create_feature function from the previous notebook
  
  Train_data_IF = Feature_pdf_IF.copy()
  Train_data_IF = Train_data_IF[features_to_test]
  #### Selecting the features on which the predictions will be made.
  
  Train_data_IF = Train_data_IF.fillna(Train_data_IF.mean())
  ## Filling the missing values with mean
  
  clf = IsolationForest(random_state=0,n_estimators = 100,contamination = 0.05).fit(Train_data_IF)
  predictions_IF = clf.predict(Train_data_IF)
    ## Training the Isolation forest model which will label all the traces into -1 or 1.
  
  
  Feature_pdf_IF['Isolation_forest_Predictions'] = predictions_IF.tolist()
  ### Storing the predictions from Isolation forest model
  
  Feature_pdf_IF['Isolation_forest_Predictions'] = Feature_pdf_IF['Isolation_forest_Predictions'].replace(-1,0)
  Feature_pdf_IF['TraceNum'] = Feature_pdf_IF['TraceNum'].astype('float')
  Feature_pdf_IF['TraceDir_New'] = Feature_pdf_IF['TraceDir_New'].astype('float')
  Feature_pdf_IF['Isolation_forest_Predictions'] = Feature_pdf_IF['Isolation_forest_Predictions'].astype('float')
  
  ## Preparing the data for training Random Forest model
  Feature_pdf_RF = Feature_pdf_IF.copy()
  Train_data_RF = Feature_pdf_RF.copy()
  Train_data_RF = Train_data_RF[features_to_test]
  Train_data_RF = Train_data_RF.fillna(Train_data_RF.mean())
  X = Train_data_RF.copy()
  y = Feature_pdf_IF[['Isolation_forest_Predictions']].values
  
  ## Training Random Forest Model
  model = RandomForestClassifier(n_estimators= 151,min_samples_split= 10,min_samples_leaf=12,max_features='auto',max_depth= 1067,criterion= 'entropy')
  model.fit(X, y)
  
  ## Saving the Random Forest Model
  pickle.dump(model, open(Model_Output_Path, 'wb'))
  
  return model
```
