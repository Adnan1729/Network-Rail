## This application is to predict on new data using the trained Random Forest model.

### This folder contains following 7 files
##### * Variables  -- This file has all the imports and variables.
##### * Driver_function -- This is the main file which will execute all the commands or files.
##### * 1.Load_Data_And_Construct_Traces  -- Here the raw data is loaded, traces as extracted and three phases of the traces are marked.
##### * 2.Feature_Creation -- This file creates features from the traces.
##### * 3.Load_Model_And_Predict -- Here the trained random forest model is loaded and is used to  get predictions at trace level using the features.
##### * 4.Convert_Prediction_To_Day_Level -- Here the trace level results are converted into day level.
##### * 5.Post_Processing_And_Final_Result -- In this file post processing step which is low pass filter is applied in the day level data and the final             result is generated
